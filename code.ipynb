```python
import pandas as pd

def filter_by_min_counts(df, user_col, item_col, min_user_reviews, min_item_reviews, max_iter=10):
    for _ in range(max_iter):
        original_shape = df.shape

        # Filtra prodotti con almeno N recensioni
        item_counts = df[item_col].value_counts()
        valid_items = item_counts[item_counts >= min_item_reviews].index
        df = df[df[item_col].isin(valid_items)]

        # Filtra utenti con almeno M recensioni
        user_counts = df[user_col].value_counts()
        valid_users = user_counts[user_counts >= min_user_reviews].index
        df = df[df[user_col].isin(valid_users)]

        # Stop se non cambia più
        if df.shape == original_shape:
            break

    return df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def create_user_item_matrix_sample(df, sample_users=100, sample_items=100):
    # Prima rimuovi i duplicati mantenendo solo l'ultima valutazione per utente-prodotto
    df_clean = df.drop_duplicates(subset=['user_id', 'parent_asin'], keep='last')
    
    # Prendi un campione di utenti e prodotti
    sample_users_list = df_clean['user_id'].unique()[:sample_users]
    sample_items_list = df_clean['parent_asin'].unique()[:sample_items]
    
    # Filtra il dataset
    sample_df = df_clean[df_clean['user_id'].isin(sample_users_list) & 
                        df_clean['parent_asin'].isin(sample_items_list)]
    
    # Crea la matrice
    matrix = sample_df.pivot(index='user_id', columns='parent_asin', values='rating')
    return matrix

# Visualizza la sparsità
matrix_sample = create_user_item_matrix_sample(df_final)

plt.figure(figsize=(12, 8))
# Crea una matrice binaria (1 se c'è rating, 0 se NaN)
binary_matrix = ~matrix_sample.isna()
sns.heatmap(binary_matrix, cbar=True, xticklabels=False, yticklabels=False, 
            cmap='Blues', cbar_kws={'label': 'Rating presente'})
plt.title(f'Visualizzazione Sparsità Matrice Utente-Prodotto\n(Campione: {len(matrix_sample)} utenti × {len(matrix_sample.columns)} prodotti)')
plt.xlabel('Prodotti')
plt.ylabel('Utenti')
plt.show()
```

```python
def create_filled_rating_matrix(df_base, algo):
    users_id = df_base["user_id"].unique()
    items_id = df_base["parent_asin"].unique()
    filled_rating_matrix = []

    print(f"Creazione matrice {len(users_id)} utenti × {len(items_id)} item...")
    
    for i, uid in enumerate(users_id):
        if i % 100 == 0:  # Progress indicator
            print(f"Processando utente {i+1}/{len(users_id)}")
            
        filled_rating_matrix.append([])
        for iid in items_id:
            res = algo.predict(uid=uid, iid=iid)
            if res.r_ui is not None:
                # Mantieni il rating originale
                filled_rating_matrix[-1].append(res.r_ui)
            else:
                # Rating predetto
                filled_rating_matrix[-1].append(res.est)

    filled_rating_matrix = np.array(filled_rating_matrix)
    print(f"Matrice creata con shape: {filled_rating_matrix.shape}")
    
    return filled_rating_matrix, users_id, items_id
```

```python
filled_rating_matrix_knn, users_id, items_id = create_filled_rating_matrix(df_base, algo)
```

```python
user_seen_items = {}
for _, row in df_base.iterrows():
    user_id = row['user_id']
    item_id = row['parent_asin']
    if user_id not in user_seen_items:
        user_seen_items[user_id] = set()
    user_seen_items[user_id].add(item_id)
```

```python
def get_clean_recommendations_wide_with_titles(filled_rating_matrix, users_id, items_id, user_seen_items, meta_df, k=10):
    rows = []
    
    for i, user_id in enumerate(users_id):
        user_ratings = filled_rating_matrix[i]
        seen_items = user_seen_items.get(user_id, set())
        
        # Solo item NON visti
        unseen_items = [(items_id[j], user_ratings[j]) 
                       for j in range(len(items_id)) 
                       if items_id[j] not in seen_items]
        
        # Ordina per rating predetto e prendi top-k
        unseen_items.sort(key=lambda x: x[1], reverse=True)
        top_k_items = unseen_items[:k]
        
        # Crea riga per il DataFrame
        row = {'user_id': user_id}
        for rank in range(k):
            if rank < len(top_k_items):
                item_id, rating = top_k_items[rank]
                
                # Cerca il titolo
                title_match = meta_df.loc[meta_df["parent_asin"] == item_id, "title"]
                title = title_match.values[0] if len(title_match) > 0 else "Title not found"
                
                row[f'rec_{rank+1}_item'] = item_id
                row[f'rec_{rank+1}_title'] = title
                row[f'rec_{rank+1}_rating'] = round(rating, 3)
            else:
                row[f'rec_{rank+1}_item'] = None
                row[f'rec_{rank+1}_title'] = None
                row[f'rec_{rank+1}_rating'] = None
        
        rows.append(row)
    
    return pd.DataFrame(rows)

# Genera versione wide
recommendations_knn = get_clean_recommendations_wide_with_titles(
    filled_rating_matrix_knn, users_id, items_id, user_seen_items, meta_sample, k=10
)
```

```python
def get_clean_recommendations_wide_with_titles(filled_rating_matrix, users_id, items_id, user_seen_items, meta_df, k=10):
    rows = []
    
    for i, user_id in enumerate(users_id):
        user_ratings = filled_rating_matrix[i]
        seen_items = user_seen_items.get(user_id, set())
        
        # Solo item NON visti
        unseen_items = [(items_id[j], user_ratings[j]) 
                       for j in range(len(items_id)) 
                       if items_id[j] not in seen_items]
        
        # Ordina per rating predetto e prendi top-k
        unseen_items.sort(key=lambda x: x[1], reverse=True)
        top_k_items = unseen_items[:k]
        
        # Crea riga per il DataFrame
        row = {'user_id': user_id}
        for rank in range(k):
            if rank < len(top_k_items):
                item_id, rating = top_k_items[rank]
                
                # Cerca il titolo
                title_match = meta_df.loc[meta_df["parent_asin"] == item_id, "title"]
                title = title_match.values[0] if len(title_match) > 0 else "Title not found"
                
                row[f'rec_{rank+1}_item'] = item_id
                row[f'rec_{rank+1}_title'] = title
                row[f'rec_{rank+1}_rating'] = round(rating, 3)
            else:
                row[f'rec_{rank+1}_item'] = None
                row[f'rec_{rank+1}_title'] = None
                row[f'rec_{rank+1}_rating'] = None
        
        rows.append(row)
    
    return pd.DataFrame(rows)

# Genera versione wide
recommendations_df = get_clean_recommendations_wide_with_titles(
    filled_rating_matrix, users_id, items_id, user_seen_items, meta_sample, k=10
)
```

```python
import ast
import re

# Funzione per convertire da lista a stringa, oppure ripulire le parentesi quadre
def clean_description(x):
    if isinstance(x, str):
        if x.startswith("["):
            try:
                # prova a valutare come lista
                items = ast.literal_eval(x)
                if isinstance(items, list):
                    return " ".join(items)
            except (ValueError, SyntaxError):
                pass  # parsing fallito, si procede sotto
        # se non è una lista valida, rimuovi manualmente parentesi quadre
        return re.sub(r'[\[\]]', '', x)
    return str(x)

# Applica la funzione
df_meta_nlp["description"] = df_meta_nlp["description"].apply(clean_description)

# Gestisci valori nulli
df_meta_nlp["title"] = df_meta_nlp["title"].fillna("")
df_meta_nlp["description"] = df_meta_nlp["description"].fillna("")

# Combina
df_meta_nlp["combined_text"] = df_meta_nlp["title"] + " " + df_meta_nlp["description"]
```

```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
```

```python
def preprocess_text_for_nlp(df, text_column, output_column='tokens', 
                           remove_stopwords=True, remove_punctuation=True, 
                           lemmatize=True, lowercase=True, verbose=False):
    """
    Preprocessa il testo per analisi NLP: tokenizzazione, rimozione stopwords, 
    punteggiatura e lemmatizzazione.
    
    Parameters:
    -----------
    df : pd.DataFrame
        DataFrame contenente i testi da processare
    text_column : str
        Nome della colonna contenente il testo da processare
    output_column : str, default='tokens'
        Nome della colonna dove salvare i token processati
    remove_stopwords : bool, default=True
        Se rimuovere le stopwords
    remove_punctuation : bool, default=True
        Se rimuovere la punteggiatura
    lemmatize : bool, default=True
        Se applicare la lemmatizzazione
    lowercase : bool, default=True
        Se convertire tutto in minuscolo
    verbose : bool, default=False
        Se stampare esempi durante il processing
    
    Returns:
    --------
    pd.DataFrame
        DataFrame con la nuova colonna contenente i token processati
    """
    
    # Crea una copia per evitare modifiche inplace
    df_processed = df.copy()
    
    # Step 1: Tokenizzazione
    if verbose:
        print("Step 1: Tokenizzazione...")
    df_processed[output_column] = df_processed[text_column].apply(word_tokenize)
    
    if verbose:
        print("\nEsempio dopo tokenizzazione:")
        for i, row in df_processed[output_column].head(3).items():
            print(f"Riga {i}: {row[:10]}...")  # Mostra solo primi 10 token
            print('-' * 40)
    
    # Step 2: Rimozione stopwords e punteggiatura
    if remove_stopwords or remove_punctuation:
        if verbose:
            print("\nStep 2: Rimozione stopwords e/o punteggiatura...")
        
        stop_words = set(stopwords.words("english")) if remove_stopwords else set()
        
        df_processed[output_column] = df_processed[output_column].apply(
            lambda tokens: [
                word for word in tokens 
                if (not remove_stopwords or (word.casefold() if lowercase else word.lower()) not in stop_words)
                and (not remove_punctuation or word not in string.punctuation)
            ]
        )
        
        if verbose:
            print("\nEsempio dopo rimozione:")
            for i, row in df_processed[output_column].head(3).items():
                print(f"Riga {i}: {row[:10]}...")
                print('-' * 40)
    
    # Step 3: Lemmatizzazione
    if lemmatize:
        if verbose:
            print("\nStep 3: Lemmatizzazione...")
        
        lemmatizer = WordNetLemmatizer()
        df_processed[output_column] = df_processed[output_column].apply(
            lambda tokens: [lemmatizer.lemmatize(word.lower() if lowercase else word) for word in tokens]
        )
        
        if verbose:
            print("\nEsempio dopo lemmatizzazione:")
            for i, row in df_processed[output_column].head(3).items():
                print(f"Riga {i}: {row[:10]}...")
                print('-' * 40)
    
    # Step 4: Lowercase (se non già applicato)
    elif lowercase:  # Solo se non è stata fatta lemmatizzazione (che già fa lowercase)
        df_processed[output_column] = df_processed[output_column].apply(
            lambda tokens: [word.lower() for word in tokens]
        )
    
    if verbose:
        print(f"\nProcessing completato! Totale righe processate: {len(df_processed)}")
    
    return df_processed
```
